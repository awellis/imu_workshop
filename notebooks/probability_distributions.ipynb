# Some probability distributions

```r
library(ggplot2)
```

## Normal distribution

[Normal distribution on Wikipedia](http://en.wikipedia.org/wiki/Normal_distribution)

```r
ggplot(data.frame(x = c(-3, 3)), aes(x)) + 
    stat_function(fun = dnorm, args = list(mean = 0, sd = 0.2)) +
    theme_bw()
```

## Binomial distribution

[Binomial distribution on Wikipedia](http://en.wikipedia.org/wiki/Binomial_distribution)

```r
N <- 20
x <- seq(0, N)
prob <- 0.6

dat <- data.frame(k = factor(0:N), prob_k = dbinom(x, N, prob))

p <- ggplot(dat, aes(x = k, y = prob_k))
p + 
#     geom_segment(aes(xend = k, yend = 0), size = 1) + 
    geom_bar(stat = "identity",  
             col = "cornflowerblue", 
             fill = "cornflowerblue") +
    ylab('p(k)') +
    scale_x_discrete(breaks=seq(0, N, 5)) +
    theme_classic()
```

## Beta distribution

[Beta distribution on Wikipedia](http://en.wikipedia.org/wiki/Beta_distribution)

```r
a <-20
b <- 20

ggplot(data.frame(x = c(0, 1)), aes(x)) + 
    stat_function(fun = dbeta, args = list(shape1 = a, shape2 = b)) +
    theme_classic()
```

## Gamma distribution

```r
ggplot(data.frame(x = c(0, 20)), aes(x)) + 
    stat_function(fun = dgamma, args = list(shape = 4, rate = 1)) +
    theme_bw()
```

## Cauchy distribution

```r
ggplot(data.frame(x = c(-10, 10)), aes(x)) + 
    stat_function(fun = dcauchy, args = list(location = 0, scale = 1)) +
    theme_bw()
```

## Student's t distribution

```r
ggplot(data.frame(x = c(-10, 10)), aes(x)) + 
    stat_function(fun = dt, args = list(df = 1)) +
    theme_bw()
```
